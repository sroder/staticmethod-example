```python
ApplicationConstants = r"C:\Users\kumarsa\Desktop\Experiment"
from gensim.models import KeyedVectors


class Vectorization():
    """
    methods:
    get_word_vec
    get_vectors
    """
    @staticmethod
    def get_word_vec():
        """
        Loads the wiki.de.vec Word Vector Fasttext wiki
        :return:
        """
        word_vectors = KeyedVectors.load_word2vec_format(ApplicationConstants + r'\wiki.de.vec')
        return word_vectors

    def get_vectors(self, tokens):
        """
        returns a vector
        look up for the vector from word_vectors matrix for the input tokens(string)
        :param tokens:
        :return:
        """

        vector_list = []
        unmatched_tokens = []
        word_vectors = self.get_word_vec()

        for token in tokens:
            try:
                vec_found = word_vectors[token]
                vector_list.append(vec_found)
            except:  # if token is out of vocab
                unmatched_tokens.append(token)

    # with open('unmatched_words.txt', 'a', encoding="utf-8") as myfile:
    # myfile.write('\n'.join(unmatched_tokens))

        return vector_list
```
